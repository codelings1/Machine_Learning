```python
# datetime.datetime.strptime("20/01/2012", format='%d/%m/%y')  this function is used finally to parse the date
# Some dates such as 23/3/20 as present in example are ambiguous. It can be 20 March 1923 or 23 March 1920/2020

import re
import datetime

def parse_date(separator):
    try:
        date_ele = date.split(separator)
        # Check if date has 3 parts i.e date, month and year else invalid date
        if(len(date_ele) == 3):
            
            total_days = 0
            month = date_ele[1]
            if(month in month_days):
            
                total_days = month_days[month]
            else:
            
                print("Invalid Date: ", date)
                return

            len_start = len(date_ele[0])
            len_last = len(date_ele[2])

            # Check [D]D-MM-YY or YY-MM-[D]D
            if((len_start == 2 and len_last <= 2) or (len_start <= 2 and len_last == 2)):
                # If both starting digits and ending digits can be day, then ambiguous
                if(0 < int(date_ele[0]) <= total_days and 0 < int(date_ele[2]) <= total_days and len_start == len_last):
                    print("Ambiguous Date: ", date, f"Day could be:{date_ele[0]} or {date_ele[2]}, similarly, mulltiple values possible for year. Need to  decide based on context.", )
                    return
                # Start digits are day
                elif(0 < int(date_ele[0]) <= total_days and len_last == 2):
                    print("Date is: ", datetime.datetime.strptime(date, f"%d{separator}%m{separator}%y"))
                # End digits are day
                elif(0 < int(date_ele[2]) <= total_days and len_start == 2):
                    print("Date is: ", datetime.datetime.strptime(date, f"%y{separator}%m{separator}%d"))
                #None are day
                else:
                    print("InCorrect Date: ", date)

            # Similar logic to above
            # [D]D-MM-YYYY or YYYY-MM-[D]D
            elif((len_start == 4 and len_last <= 2) or (len_start <= 2 and len_last == 4)):
                if(0 < int(date_ele[0]) <= total_days and len_last == 4):
            
                    print("Date is: ", datetime.datetime.strptime(date, f"%d{separator}%m{separator}%Y"))
                elif(0 < int(date_ele[2]) <= total_days and len_start == 4):
            
                    print("Date is: ", datetime.datetime.strptime(date, f"%Y{separator}%m{separator}%d"))
                else:
                    print("InCorrect Date: ", date)
            else:
                print("Invalid Date: ", date)


        else:
            
            print("Invalid Date: ", date)
    except Exception as e:
        print("Invalid Date", date)

        
        
text = ""
with open('./date_parse.txt', 'r') as read_file:
    text = read_file.read() 

# Regex to extract dates from the document . Currently these kind are handled ==> # [D]D-MM-YY or YY-MM-[D]D or # [D]D-MM-YYYY or YYYY-MM-[D]D
date_regex = "[0-9]?[0-9]?[0-9]?[0-9][\/-][0-1]?[0-9][\/-][0-9]?[0-9]?[0-9][0-9]"

# Days in a month dict
month_days = {'01': 31,'02': 29,'03': 31,'04': 30,'05': 31,'06': 30,'07': 31,'08': 31,
             '09': 30,'10': 31,'11': 30,'12': 31,
             '1': 31,'2': 29,'3': 31,'4': 30,'5': 31,'6': 30,'7': 31,'8': 31,
             '9': 30,'10': 31,'11': 30,'12': 31}

# Check for a leap year is done automatically by datetime.datetime

# Extract all date like patterns from the document
dates = re.findall(date_regex, text)

print(dates)
# Iterate 
for date in dates:
    if("-" in date):
        parse_date("-")        
    elif("/" in date):
        parse_date("/")
    else:
        print("Invalid Date: ", date)
```
