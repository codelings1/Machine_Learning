```python
def df_convert_row_to_column(df):
    base_cols = ['session_id', 'startTime', 'endTime', 'gender'] if 'gender' in df.columns else \
    ['session_id', 'startTime', 'endTime']
    
    df1 = df.copy()
    df1['product_list'] = df1['ProductList'].str.split(';')
    df1['products_count'] = df1['product_list'].str.len()
    
    df2 = pd.DataFrame({col:np.repeat(df1[col].values, df1['products_count'].values) \
                             for col in base_cols})

    df_temp = pd.DataFrame(pd.Series(np.concatenate(df1['product_list'].values)).str.split('/').values.tolist())
    df_temp.columns = ['cat', 'subcat', 'subcat2', 'product', 'todrop']
    df_temp.drop(columns=['todrop'], inplace=True)

    df2 = pd.concat([df2, df_temp], axis=1)

    start_time = pd.to_datetime(df2['startTime'])
    df2['st_date'] = start_time.dt.day
    df2['st_day'] = start_time.dt.weekday
    df2['st_hour'] = start_time.dt.hour
    df2['st_minute'] = start_time.dt.minute
    df2['st_weekday'] = start_time.dt.weekday
    df2['st_weekofyear'] = start_time.dt.weekofyear
    df2['st_month'] = start_time.dt.month
    df2['st_is_month_end'] = start_time.dt.is_month_end

    end_time = pd.to_datetime(df2['endTime'])
    df2['et_date'] = end_time.dt.day
    df2['et_day'] = end_time.dt.weekday
    df2['et_hour'] = end_time.dt.hour
    df2['et_minute'] = end_time.dt.minute
    df2['et_weekday'] = end_time.dt.weekday
    df2['et_weekofyear'] = end_time.dt.weekofyear
    df2['et_month'] = end_time.dt.month
    df2['et_is_month_end'] = end_time.dt.is_month_end
    
    df2['duration'] = (end_time - start_time).dt.seconds
    
    return df2.copy()
```
